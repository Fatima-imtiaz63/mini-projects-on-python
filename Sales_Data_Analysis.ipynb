```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
import os
df=pd.read_csv('sales_data_analysis-1.csv')
df
df.head()
df.info()
df.isnull().sum()
df=df.dropna()
df=df.drop_duplicates()
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df
df.shape
df.describe()
df['Region'].value_counts()
df.groupby('Region')['Sales'].sum()
df.corr(numeric_only=True)
df.groupby('Region')['Sales'].sum().plot()
df.groupby('Category')['Sales'].sum()
df.groupby('Category')['Sales'].sum().plot()
df['Month']=df['Order_Date'].dt.month
df
monthly_sales=df.groupby('Month')['Sales'].sum()
monthly_sales.plot()
region_sales=df.groupby('Region')['Sales'].sum()
region_sales.plot()
plt.figure(figsize=(8,5))

plt.plot(monthly_sales.index, monthly_sales.values, marker='o', label='Sales Trend')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.legend()
category_sales=df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title("Category Sales Comparison")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
category_sales.plot()
region_sales.plot(kind='bar')
plt.show()
```
